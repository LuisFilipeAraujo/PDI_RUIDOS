```python
# Importação das bibliotecas
import numpy as np
import argparse
import cv2 
from matplotlib import pyplot as plt
from google.colab import files

from skimage.util import random_noise
```

```python
uploaded = files.upload()
```

```python
img = cv2.imread('goku.jpg')
```

```python
RGB_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

## **RUÍDO**

```python
I = RGB_img
gauss = random_noise(I, mode='gaussian', seed=None, clip=True)
sp = random_noise(I, mode='s&p', seed=5, clip=True)

fig1 = plt.figure(figsize=(20,50))
ax1 = fig1.add_subplot(131)
plt.imshow(I,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Original')

ax2 = fig1.add_subplot(132)
plt.imshow(gauss,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Gaussian')

ax3 = fig1.add_subplot(133)
plt.imshow(sp,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Salt & Pepper')

plt.show()
```

```python
imgNC = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
```

```python
I1 = imgNC
gauss = random_noise(I1, mode='gaussian', seed=None, clip=True)
sp = random_noise(I1, mode='s&p', seed=5, clip=True)

fig1 = plt.figure(figsize=(20,50))
ax1 = fig1.add_subplot(131)
plt.imshow(I1,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Original')

ax2 = fig1.add_subplot(132)
plt.imshow(gauss,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Gaussian')

ax3 = fig1.add_subplot(133)
plt.imshow(sp,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Salt & Pepper')

plt.show()
```

## **FILTRO BLUR**

```python
blur = cv2.blur(I,(10,10))

fig3 = plt.figure(figsize=(15,20))
ax1 = fig3.add_subplot(121)
plt.imshow(I),plt.title('Original')
plt.xticks([]), plt.yticks([])

ax2 = fig3.add_subplot(122)
plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
blur1 = cv2.blur(I1,(10,10))

fig4 = plt.figure(figsize=(15,20))
ax1 = fig4.add_subplot(121)
plt.imshow(I1,'gray'),plt.title('gray')
plt.xticks([]), plt.yticks([])

ax2 = fig4.add_subplot(122)
plt.imshow(blur1,'gray'),plt.title('Gray Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
```
